```python
#let us do binning for the age of people applying for loan.
#as we have the DAYS_BIRTH column containing all -ve values, let us see if there are outliers
sum(appdata.DAYS_BIRTH>0)
```

```python
# creating a bins for various ages
def age_period(days):
    days=abs(days)
    x=int(days/365)
    'divide the age into 3 categories'
    if 0< x < 30:
        return "Young Age"
    elif 30 <= x < 60:
        return "Middle Age"
    else:
        return "Old age"

appdata['Age_Group'] = appdata.DAYS_BIRTH.apply(lambda x: age_period(x))
appdata.Age_Group.value_counts()
```

```python
sns.catplot(x = "Age_Group",data = appdata, kind = "count")
```

Maximum people applying for the loan are middle age i.e; 30 to 60years old.
Let us further bin them into age groups

```python
def age(days):
    days=abs(days)
    x=int(days/365)
    return x
appdata['Age'] = appdata.DAYS_BIRTH.apply(lambda x: age(x))   
print(appdata.Age.min())
print(appdata.Age.max())
appdata['age_bins'] = pd.cut(x=appdata['Age'], bins=[20, 29, 39, 49,59,69])
sns.catplot(x = "age_bins",data = appdata, kind = "count")
#max people applying for loans are in their 30s
```

```python
# creating a time slots for various time period of day
def time_period(x):
    'divide the time of the day into five categories'
    if x < 5:
        return "Early_Morning"
    elif 5 <= x < 10:
        return "Morning"
    elif 10 <= x < 17:
        return "Day_Time"
    elif 17 <= x < 22:
        return "Evening"
    else:
        return "Late_Night"
appdata['Time_slot'] = appdata.HOUR_APPR_PROCESS_START.apply(lambda x: time_period(x))
g=sns.catplot(x = "Time_slot",data = appdata, kind = "count",order=['Early_Morning','Morning','Day_Time','Evening','Late_Night'])
#we notice max applications applying during 10 to 5(day time)
```

```python
#let us see another column
appdata.AMT_INCOME_TOTAL.describe()
```

```python
# creating a bins for income
def income(x):
    'divide the income into 3 categories'
    if 0< x < 135000:
        return "Low Income"
    elif 135000 <= x < 300000:
        return "Average Income"
    else:
        return "High income"

appdata['Income_Bin'] = appdata.AMT_INCOME_TOTAL.apply(lambda x: income(x))
sns.catplot(x = "Income_Bin",data = appdata, kind = "count")
```

Maximum people applying for loan have low income

Correlation determination :

```python
#for correlation lets go ahead and drop some columns
app0=appzero.drop(['SK_ID_CURR', 'TARGET', 'NAME_CONTRACT_TYPE', 'CODE_GENDER','FLAG_OWN_CAR', 'FLAG_OWN_REALTY','NAME_TYPE_SUITE',
       'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS','DAYS_BIRTH',
       'OCCUPATION_TYPE','WEEKDAY_APPR_PROCESS_START',
       'HOUR_APPR_PROCESS_START', 'ORGANIZATION_TYPE','Age_Group',
       'age_bins', 'Time_slot', 'Income_Bin'],axis=1)
```

```python
drop_zero= app0.corr(method='pearson')
drop_zero
```

```python
sns.heatmap(drop_zero, xticklabels=drop_zero.columns,yticklabels=drop_zero.columns, cmap="Blues")
plt.rcParams["figure.figsize"] = (10,7)
```

```python
#let us find the top 10 correlations
drop_zero.corr().unstack().sort_values().drop_duplicates()
```

We notice the top 10 correlations from this output.

The top 5 positively correlated pais are:

AMT_ANNUITY                  AMT_GOODS_PRICE               
AMT_CREDIT                   AMT_ANNUITY                
CNT_FAM_MEMBERS              CNT_CHILDREN               
REGION_RATING_CLIENT_W_CITY  REGION_RATING_CLIENT       
AMT_GOODS_PRICE              AMT_CREDIT 

Which means if one of the pair value increases, so does the other.

The top 5 negatively correlated pairs are:

REGION_RATING_CLIENT         REGION_POPULATION_RELATIVE    
REGION_POPULATION_RELATIVE   REGION_RATING_CLIENT_W_CITY   
Age                          CNT_CHILDREN
Age                          CNT_FAM_MEMBERS               
REGION_RATING_CLIENT_W_CITY  AMT_INCOME_TOTAL 

Which means if one of the pair value increases, the other value decreases.
