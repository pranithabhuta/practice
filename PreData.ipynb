```python
predf.NAME_YIELD_GROUP=predf.NAME_YIELD_GROUP.astype('category')
predf.NAME_CONTRACT_TYPE=predf.NAME_CONTRACT_TYPE.astype('category')
predf.WEEKDAY_APPR_PROCESS_START=predf.WEEKDAY_APPR_PROCESS_START.astype('category')
predf.dtypes
#we will be using these categorical values for our univariate and bi-variate analysis
```

```python
def years(days):
    days=abs(days)
    x=int(days/365)
    return x
predf['YEARS_TERMINATION'] = predf.DAYS_TERMINATION.apply(lambda x: years(x))   
predf['YEARS_FIRST_DUE'] = predf.DAYS_FIRST_DUE.apply(lambda x: years(x))   
predf['YEARS_DECISION']=predf.DAYS_DECISION.apply(lambda x:years(x))
#so that we can compare in terms of years.
```

```python
g=sns.catplot(x = "WEEKDAY_APPR_PROCESS_START",hue='NAME_CONTRACT_TYPE',data = predf, kind = "count",order=['MONDAY','TUESDAY','WEDNESDAY','THURSDAY','FRIDAY','SATURDAY','SUNDAY'])
g.fig.set_size_inches(12,5)
#Sunday cash loans were the were least
#Saturday Consumers loans were the highest
```

```python
# creating a time slots for various time period of day
def time_period(x):
    'divide the time of the day into five categories'
    if x < 5:
        return "Early_Morning"
    elif 5 <= x < 10:
        return "Morning"
    elif 10 <= x < 17:
        return "Day_Time"
    elif 17 <= x < 22:
        return "Evening"
    else:
        return "Late_Night"
predf['Time_slot'] = predf.HOUR_APPR_PROCESS_START.apply(lambda x: time_period(x))
g=sns.catplot(x = "Time_slot",hue='WEEKDAY_APPR_PROCESS_START',data = predf, kind = "count",order=['Early_Morning','Morning','Day_Time','Evening','Late_Night'])
#we can conlude that evening is the only time on saturday when applicants are not the higher when compared to other days.
#We can say that the applications start reducing from saturday eve and remain low on Sunday as seen before
```
